```python
from sklearn.datasets import load_iris
iris_data = load_iris()
```

```python
iris_data
```

```python
import pandas as pd
data_ip=pd.DataFrame.from_dict(iris_data['data'])
```

```python
data_ip1=data_ip.rename(columns = {0: 'sepal_length', 1: 'sepal_width',2:'petal_length',3:'petal_width'},inplace=False)
```

```python
data_ip1
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=data_ip1,hue=iris_data.target)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=data_ip1,color='olive')
plt.show() # scatter plot and histogram plot for sepal length  at the top and corresponding histogram plot for petal length on the right
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=data_ip1,kind='reg',color='olive')
plt.show() # scatter plot and histogram plot for sepal length  at the top and corresponding histogram plot for petal length on the right
```
